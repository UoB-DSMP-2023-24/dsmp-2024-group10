```python
import pandas as pd 
df= pd.read_csv("vdjdb.txt", sep="\t")
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 7)
df.to_csv('out.csv', index=False) 
df=pd.read_csv('out.csv')
df=df.drop(columns=["reference.id","method","meta","cdr3fix","web.method","web.method.seq","web.cdr3fix.nc","web.cdr3fix.unmp"])
```

```python
df_A=df[(df["gene"] == "TRA") & (df["vdjdb.score"] > 0)]
df_B=df[(df["gene"] == "TRB") & (df["vdjdb.score"] > 0)]
df_AB=df[df["complex.id"] > 1]
```

```python
df1=df_AB.groupby('complex.id', as_index=False).agg(lambda x: set(x))
```

```python
split = pd.DataFrame(df1["cdr3"].to_list(), columns = ['cdrA', 'cdrB'])
df1 = pd.concat([df1, split], axis=1)
df1["combined"] = df1["cdrA"].astype(str) + df1["cdrB"]
split= pd.DataFrame(df1["v.segm"].to_list(), columns = ['v.segmA', 'v.segmB'])
df1 = pd.concat([df1, split], axis=1)
```

```python
df_AG=df_A[ ['cdr3', 'v.segm'] + [c for c in df_A if c not in ['cdr3', 'v.segm']] ]
df_BG=df_B[ ['cdr3', 'v.segm'] + [c for c in df_B if c not in ['cdr3', 'v.segm']] ]
df1=df1[ ['combined', 'v.segmA'] + [c for c in df_B if c not in ['cdr3', 'v.segmA']] ]
df_AG=df_AG.rename(columns={"cdr3":"aminoAcid","v.segm":"vMaxResolved"})
df_BG=df_BG.rename(columns={"cdr3":"aminoAcid","v.segm":"vMaxResolved"})
```

```python
d=df1
for c in d.columns:

    d[c]= d.apply(lambda x: str(x[c]).replace("{",'').replace('}','').strip(" ' "), axis=1)
```

```python
df_AG.to_csv('df_AG.txt', sep="\t", index=False)
df_BG.to_csv('df_BG.txt', sep="\t", index=False)
d.to_csv('df_AB.txt', sep="\t", index=False)
```
